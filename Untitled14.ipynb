```python
# Instalando as dependências necessárias
!pip install yt-dlp
!pip install git+https://github.com/openai/whisper.git
!pip install gTTS
!pip install openai==0.28
!pip install gradio
!pip install edge-tts
!apt install ffmpeg -y

import subprocess
import os
import asyncio
import whisper
import openai
from gtts import gTTS
import gradio as gr
import edge_tts  # para a nova síntese de voz

# Defina sua chave de API do OpenAI
openai.api_key = "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"

def ler_vtt(vtt_file):
    """
    Lê arquivo VTT e retorna o texto extraído (ignorando timestamps).
    """
    texto = ""
    with open(vtt_file, "r", encoding="utf-8") as f:
        for linha in f:
            linha = linha.strip()
            # Ignora cabeçalhos e linhas de tempo
            if linha and not (linha.startswith("WEBVTT") or "-->" in linha):
                texto += linha + " "
    return texto.strip()

def baixar_e_transcrever(youtube_url):
    """
    Baixa o áudio do vídeo do YouTube, tenta obter legendas automáticas e,
    se não encontradas, usa o Whisper para transcrição.
    """
    # Remove arquivos antigos (se existirem)
    for arquivo in ["audio.m4a", "audio.wav", "auto_sub.vtt"]:
        if os.path.exists(arquivo):
            os.remove(arquivo)

    # Tenta baixar as legendas automáticas em português
    comando_subs = [
        "yt-dlp", youtube_url,
        "--write-auto-sub", "--sub-lang", "pt",
        "--skip-download",
        "-o", "auto_sub.%(ext)s"
    ]
    resultado_subs = subprocess.run(comando_subs, capture_output=True, text=True)

    if os.path.exists("auto_sub.vtt"):
        # Se legendas foram baixadas, usa-as
        transcription = ler_vtt("auto_sub.vtt")
        if transcription:
            print("Transcrição obtida via legendas do YouTube.")
            return transcription

    # Se não houver legendas, baixa o áudio no formato m4a
    comando_download = ["yt-dlp", youtube_url, "--format", "m4a", "-o", "audio.m4a"]
    subprocess.run(comando_download, check=True)

    # Converte para WAV usando ffmpeg
    comando_ffmpeg = ["ffmpeg", "-i", "audio.m4a", "-c:a", "pcm_s16le", "audio.wav", "-y"]
    subprocess.run(comando_ffmpeg, check=True)

    # Carrega o modelo "tiny" para uma transcrição mais rápida
    model = whisper.load_model("tiny")
    result = model.transcribe("audio.wav", language="portuguese")
    transcription = result["text"]
    return transcription

def gerar_perguntas(transcription):
    """
    Usa a API do GPT para gerar até 5 perguntas relevantes baseadas na transcrição do vídeo.
    """
    prompt = f"""
Baseado no seguinte conteúdo do vídeo:
"{transcription}"
Gere até 5 perguntas que uma pessoa poderia responder após assistir ao vídeo.
As perguntas devem ser relevantes e cobrir os pontos principais discutidos.
"""
    response = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "Você é um assistente útil que gera perguntas com base em textos."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=200
    )
    questions = response.choices[0].message.content.strip()
    return questions

def avaliar_respostas(transcription, questions, respostas):
    """
    Avalia as respostas fornecidas pelo usuário, comparando com o conteúdo do vídeo.
    """
    prompt = f"""
Considere o seguinte conteúdo do vídeo:
"{transcription}"
E as seguintes perguntas e respostas fornecidas pelo usuário:
Perguntas:
{questions}
Respostas:
{respostas}
Para cada pergunta, avalie se a resposta corresponde ao que foi discutido no vídeo, justificando brevemente.
"""
    response = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "Você é um assistente que avalia respostas baseadas no conteúdo de um vídeo."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=300
    )
    avaliacao = response.choices[0].message.content.strip()
    return avaliacao

async def text_to_audio_openai(text, filename="output.mp3", voice="pt-BR-FranciscaNeural"):
    """
    Converte texto em áudio usando edge-tts (simulando a 'voz da OpenAI').
    """
    communicate = edge_tts.Communicate(text, voice=voice)
    await communicate.save(filename)
    return filename

def text_to_audio(text, filename="output.mp3", voice_type="gtts"):
    """
    Converte um texto em áudio.
    Se voice_type == "gtts": utiliza gTTS.
    Se voice_type == "openai": utiliza a função edge-tts para simular a voz da OpenAI.
    """
    if voice_type == "openai":
        try:
            return asyncio.run(text_to_audio_openai(text, filename=filename))
        except Exception as e:
            print("Erro ao usar voz OpenAI (edge-tts):", e)
            # fallback para gTTS
    # Padrão: usa gTTS
    tts = gTTS(text=text, lang='pt')
    tts.save(filename)
    return filename

def processar_video(youtube_url):
    """
    Processa o vídeo, baixando o áudio (ou legendas), gerando a transcrição e criando as perguntas.
    Sempre recria o state para garantir que seja usado o conteúdo do novo link.
    """
    try:
        transcricao = baixar_e_transcrever(youtube_url)
    except Exception as e:
        # Retorna um state vazio para garantir que não haja dados antigos
        return {"transcricao": "", "perguntas": ""}, "Erro ao processar o vídeo: " + str(e), ""

    try:
        perguntas = gerar_perguntas(transcricao)
    except Exception as e:
        return {"transcricao": "", "perguntas": ""}, "Erro ao gerar perguntas: " + str(e), ""

    # Gera áudio com as perguntas usando a voz "openai" (edge-tts) – troque para "gtts" se preferir
    audio_perguntas = text_to_audio(perguntas, filename="perguntas.mp3", voice_type="openai")

    # Retorna um novo state contendo a transcrição e as perguntas
    state = {"transcricao": transcricao, "perguntas": perguntas}
    return state, perguntas, audio_perguntas

def processar_respostas(state, respostas):
    """
    Avalia as respostas do usuário com base no state (novo conteúdo do vídeo).
    Gera também uma versão em áudio da avaliação.
    """
    if not state or not state.get("transcricao"):
        return "Erro: nenhum vídeo processado. Por favor, processe um vídeo primeiro.", ""
    transcricao = state.get("transcricao", "")
    perguntas = state.get("perguntas", "")
    try:
        avaliacao = avaliar_respostas(transcricao, perguntas, respostas)
    except Exception as e:
        return "Erro ao avaliar respostas: " + str(e), ""

    # Converte a avaliação para áudio usando a voz "openai"
    audio_avaliacao = text_to_audio(avaliacao, filename="avaliacao.mp3", voice_type="openai")
    return avaliacao, audio_avaliacao

# Construindo a interface usando Gradio Blocks
with gr.Blocks() as demo:
    gr.Markdown("# Treine seu Aprendizado!")

    # Estado único para armazenar os dados do vídeo processado
    state_video = gr.State()

    with gr.Tab("Processar Vídeo"):
        gr.Markdown("Insira a URL do YouTube para processar o áudio/legenda e gerar perguntas sobre o conteúdo do vídeo.")
        url_input = gr.Textbox(label="Link do YouTube", placeholder="https://youtu.be/...")
        btn_processar = gr.Button("Processar Vídeo")
        # Exibe as perguntas geradas em texto e também o áudio correspondente
        perguntas_box = gr.Textbox(label="Perguntas Geradas", lines=8)
        audio_perguntas_out = gr.Audio(label="Áudio das Perguntas", type="filepath")

        # Ao clicar, a função retorna: [state, texto das perguntas, arquivo de áudio das perguntas]
        btn_processar.click(fn=processar_video, inputs=url_input, outputs=[state_video, perguntas_box, audio_perguntas_out])

    with gr.Tab("Avaliar Respostas"):
        gr.Markdown("Responda às perguntas abaixo (cada resposta em uma nova linha) e clique em 'Avaliar Respostas'.")
        # Exibe as perguntas geradas (atualizadas com base no state)
        perguntas_exibidas = gr.Textbox(label="Perguntas", interactive=False)
        respostas_input = gr.Textbox(label="Suas Respostas (uma por linha)", lines=8)
        btn_avaliar = gr.Button("Avaliar Respostas")
        avaliacao_box = gr.Textbox(label="Avaliação", lines=8)
        audio_avaliacao_out = gr.Audio(label="Áudio da Avaliação", type="filepath")

        # Atualiza a caixa de perguntas com base no state (garante que o conteúdo seja sempre do último vídeo processado)
        state_video.change(lambda s: s.get("perguntas", "") if s else "", inputs=state_video, outputs=perguntas_exibidas)

        # Ao clicar, usa o state armazenado e as respostas do usuário
        btn_avaliar.click(fn=processar_respostas, inputs=[state_video, respostas_input], outputs=[avaliacao_box, audio_avaliacao_out])

demo.launch()
```
